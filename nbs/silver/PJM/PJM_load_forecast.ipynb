```python
import pandas as pd
from usagrid import s3

df_forecast = s3.read_pyarrow_from_s3(bucket_name="usagrid",key="silver/PJM/PJM_load_forecast.arrow").to_pandas().query('forecast_area == "RTO"')

df_metered = s3.read_pyarrow_from_s3(bucket_name="usagrid",key="silver/PJM/PJM_load_metered.arrow").to_pandas().reset_index().query('load_area == "RTO"')
```

```python
rto_forecast_metered = df_metered.merge(df_forecast,left_on=["datetime_beginning_utc","load_area"],
                               right_on=["forecast_hour_beginning_utc","forecast_area"],
                               how="left",
                           indicator=True
                              )
```

```python
rto_forecast_metered["forecast_metered_pct"] = rto_forecast_metered["forecast_load_mw"]/rto_forecast_metered["mw"]

rto_forecast_metered["forecast_metered_diff"] = rto_forecast_metered["forecast_load_mw"] - rto_forecast_metered["mw"]

rto_forecast_metered["missed_forecast"] = rto_forecast_metered["forecast_metered_pct"] <= 0.95
```
